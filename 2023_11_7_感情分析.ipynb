```python
#https://qiita.com/y_itoh/items/4693bd8f64ac811f8524
#上記のサイトを参考にした

#名詞の分析の前処理
import pandas as pd
# カラム名と値の位置ずれを制御
pd.set_option('display.unicode.east_asian_width', True)

pndic_1_1 = pd.read_csv('pn.csv.m3.120408.trim', names=['word_pn_oth'])
print(pndic_1_1)
```

```python
pndic_2_1 = pndic_1_1['word_pn_oth'].str.split('\t', expand=True)
print(pndic_2_1)
```

```python
senti_score = pd.Series(pndic_2_1[1])
senti_score.value_counts()
```

```python
pndic_3_1 = pndic_2_1[(pndic_2_1[1] == 'p') | (pndic_2_1[1] == 'e') | (pndic_2_1[1] == 'n')]
print(pndic_3_1)
```

```python
#動詞形容詞の分析の前処理
pndic_1_2 = pd.read_csv('wago.121808.pn', names=['word_pn_oth'])
print(pndic_1_2)
```

```python
pndic_2_2 = pndic_1_2['word_pn_oth'].str.split('\t', expand=True)
print(pndic_2_2)
```

```python
pndic_2_2 = pndic_2_2.reindex(columns=[1, 0])
pndic_2_2 = pndic_2_2.rename(columns={1: 0, 0: 1})
print(pndic_2_2)
```

```python
senti_score = pd.Series(pndic_2_2[1])
senti_score.value_counts()
```

```python
pndic_3_2 = pndic_2_2[(pndic_2_2[1] == 'ネガ（評価)') | (pndic_2_2[1] == 'ポジ（評価）') | (pndic_2_2[1] == 'ネガ（経験）') | (pndic_2_2[1] == 'ポジ（経験）')]
print(pndic_3_2)
```

```python
pndic_4_2 = pndic_3_2
```

```python
import MeCab
mecab = MeCab.Tagger("-Ochasen")

# 1行目の形態素解析結果を例示
print(mecab.parse(lines[0]))
```

```python
word_list = []
for l in lines:
    temp = []
    for v in mecab.parse(l).splitlines():
        if len(v.split()) >= 3:
            if v.split()[3][:2] in ['名詞','形容詞','動詞','副詞']:
                temp.append(v.split()[2])
    word_list.append(temp)

# 空の要素を削除
word_list = [x for x in word_list if x != []]
```

```python
print(word_list)
```

```python
result = []
# 文単位の処理
for sentence in word_list:
    temp = []
    # 語単位の処理
    for word in sentence:
        word_score = []
        score = dic.get(word)
        word_score = (word, score)
        temp.append(word_score)
    result.append(temp)

# 文毎にデータフレーム化して表示
for i in range(len(result)):
    print(lines[i], '\n', pd.DataFrame(result[i], columns=["word", "score"]), '\n')
```

```python
# 文単位の平均値を計算
mean_list = []
num_of_sent = 0
for i in result:
    num_of_sent = num_of_sent + 1
    temp = []
    for j in i:
        if not j[1] == None:
            temp.append(float(j[1]))
    if not len(temp) == 0:
        mean = (sum(temp) / len(temp))
        mean_list.append(mean)
    else:
        mean = 0
        mean_list.append(mean)

# データフレーム化して表示
print(pd.DataFrame(mean_list, columns=["mean"], index=lines[0:num_of_sent]))
```
